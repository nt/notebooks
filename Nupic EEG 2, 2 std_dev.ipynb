```python
import os, sys
import scipy.io
import numpy as np
```

```python
import sklearn.preprocessing

class DataSet:
    def __init__(self, location, downsample_factor = 4):
        interictal = scipy.io.loadmat(location)
        seg_name =  [k for k in interictal.keys() if k not in ["__version__", '__header__', '__globals__']][0]
        data, data_length_sec, sampling_frequency, channels, sequence = interictal[seg_name][0][0]
        self.data_length_sec = data_length_sec
        self.sampling_frequency = sampling_frequency[0][0] / downsample_factor 
        channels_clean = []
        for c in channels[0]:
            channels_clean.append(c[0])
        self.channels = channels_clean
        self.sequence = sequence
        down_sampled = []
        i = 0
        for row in data.T:
            i += 1
            if i % 4 == 0:
                down_sampled.append(row)
        down_sampled = np.array(down_sampled, dtype="f").T
        whitened = []
        for channel in down_sampled:
            whitened.append(sklearn.preprocessing.scale(channel))
        self.data = np.array(whitened)
    
    def dataForSeconds(self, second_start, seconds_end = None):
        i = int(second_start * self.sampling_frequency)
        if seconds_end:
            j = int(seconds_end * self.sampling_frequency)
        else:
            j = self.data.shape[1]
        return self.data.T[:][i:j].T
    
    def __str__(self):
        length = self.data.shape[1]/self.sampling_frequency
        return "Sampling frequency: %s\nDataset duration: %ss\nChannels : %s" % (self.sampling_frequency, length, self.channels)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def plotEEG(dataset, seconds_start = 0, seconds_end = 10):
    fig, axes = plt.subplots(nrows=16, ncols=1, figsize=(25, 13))
    fig.subplots_adjust(hspace=0)
    plt.setp([a.get_xticklabels() for a in fig.axes[:-1]], visible=False)

    for i in range(0, 16):
        axes[i].plot(dataset.dataForSeconds(seconds_start, seconds_end)[i])
        axes[i].set_title(dataset.channels[i])

plotEEG(preictal)
```

```python
from nupic.data.datasethelpers import findDataset
from nupic.frameworks.opf.modelfactory import ModelFactory
from nupic.frameworks.opf.predictionmetricsmanager import MetricsManager

inter_model = ModelFactory.create(params(interictal))
anomaly_interictal = [inter_model.run({"channel1": r}).inferences['anomalyScore'] for r in interictal.dataForSeconds(0, 60)[0]]
inter_model.disableLearning()
anomaly_interictal += [inter_model.run({"channel1": r}).inferences['anomalyScore'] for r in interictal.dataForSeconds(60, 90)[0]]
```

```python
anomaly_preictal = [inter_model.run({"channel1": r}).inferences['anomalyScore'] for r in preictal.dataForSeconds(0, 90)[0]]
```

```python
def runningMean(x, N):
    y = np.zeros((len(x),))
    for ctr in range(len(x)):
         y[ctr] = np.sum(x[ctr:(ctr+N)])
    return y/N
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(25, 14))
fig.subplots_adjust(hspace=0)
plt.setp([a.get_xticklabels() for a in fig.axes[:-1]], visible=False)
K = 100

axes[0].plot(interictal.dataForSeconds(0, 90)[0])
axes[0].set_title("interictal data")
axes[1].plot(runningMean(anomaly_interictal, K))
axes[1].set_title("interictal anomaly")

axes[2].plot(preictal.dataForSeconds(0, 90)[0])
axes[2].set_title("preictal data")
axes[3].plot(runningMean(anomaly_preictal,K))
axes[3].set_title("preictal anomaly")

```
